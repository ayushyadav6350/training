```python
upper_limit=tip['total_bill'].quantile(0.9)
```

```python
sns.boxplot(tip[tip['total_bill']<upper_limit]['total_bill'])            # outliers removed
```

```python
### Using IQR range
q1=np.percentile(tip['total_bill'],25)
q3=np.percentile(tip['total_bill'],75)
iqr=q3-q1

upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
```
