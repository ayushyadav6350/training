# Matplotlib bars

```python
x= np.array(["A","B","C","D"])      # bars plot between integer and string
y= np.array([1,2,3,4])
plt.bar(x,y)
plt.show()
```

```python
x= np.array(["A","B","C","D"])      # bars plot between integer and string
y= np.array([1,2,3,4])
plt.barh(x,y)
plt.show()
```

```python
x= np.array(["A","B","C","D"])      # bars plot between integer and string
y= np.array([1,2,3,4])
plt.bar(x,y,color="orange")
plt.show()
```

# matplotlib Piecharts

```python
y= np.array([10,20,30,40])
plt.pie(y)
plt.show()
```

```python
y= np.array([10,20,30,40])
mylabels=["apple","orange","banana","grapes"]
plt.pie(y, labels= mylabels)
plt.show()
```

```python
y= np.array([10,20,30,40])
mylabels=["apple","orange","banana","grapes"]
plt.pie(y, labels= mylabels,startangle=300)
plt.show()
```

```python
y= np.array([10,20,30,40])
mylabels=["apple","orange","banana","grapes"]
myexlode=[0.2,0,0,0]
plt.pie(y, labels= mylabels,startangle=300,explode=myexlode)
plt.show()
```

```python
y= np.array([10,20,30,40])
mylabels=["apple","orange","banana","grapes"]
myexlode=[0.2,10,40,0]
plt.pie(y, labels= mylabels,startangle=300,explode=myexlode)
plt.show()
```

```python
y= np.array([10,20,30,40])
mylabels=["apple","orange","banana","grapes"]
myexlode=[0.2,0,0,0]
plt.pie(y, labels= mylabels,startangle=300,explode=myexlode,shadow=True)
plt.show()
```

```python
y= np.array([10,20,30,40])
mylabels=["apple","orange","banana","grapes"]
mycolors=["black","pink","g","grey"]
plt.pie(y, labels= mylabels, colors=mycolors)
plt.show()
```

```python
y= np.array([10,20,30,40])
mylabels=["apple","orange","banana","grapes"]
mycolors=["black","pink","g","grey"]
plt.pie(y, labels= mylabels, colors=mycolors)
plt.legend()
plt.show()
```

```python
y= np.array([10,20,30,40])
mylabels=["apple","orange","banana","grapes"]
mycolors=["black","pink","g","grey"]
plt.pie(y, labels= mylabels, colors=mycolors)
plt.legend(title="fruites names")
plt.show()
```
